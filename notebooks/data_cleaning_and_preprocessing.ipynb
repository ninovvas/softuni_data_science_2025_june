```python
# Remove the two columns
asthma_disease_df = asthma_disease_df.drop(columns = ["DoctorInCharge", "PatientID"])
asthma_disease_df
```

```python
asthma_disease_df.info()
```

```python
asthma_disease_df['BMI_Category'] = pd.cut(asthma_disease_df['BMI'], 
                                   bins=[0, 18.5, 25, 30, float('inf')],
                                   labels=['Underweight', 'Normal', 'Overweight', 'Obese'])
```

```python
asthma_disease_df
```

```python
# Create lung function ratio
asthma_disease_df['FEV1_FVC_Ratio'] = asthma_disease_df['LungFunctionFEV1'] / asthma_disease_df['LungFunctionFVC']
```

```python
asthma_disease_df
```

```python
# Create symptom severity score
symptom_cols = ['Wheezing', 'ShortnessOfBreath', 'ChestTightness', 
                'Coughing', 'NighttimeSymptoms', 'ExerciseInduced']
asthma_disease_df['SymptomSeverity'] = asthma_disease_df[symptom_cols].sum(axis=1)
```

```python
asthma_disease_df
```

```python
# Create allergy composite score
allergy_cols = ['PetAllergy', 'HistoryOfAllergies', 'Eczema', 'HayFever']
asthma_disease_df['AllergyScore'] = asthma_disease_df[allergy_cols].sum(axis=1)
```

```python
asthma_disease_df
```

```python

# Create exposure composite score
exposure_cols = ['PollutionExposure', 'PollenExposure', 'DustExposure']
asthma_disease_df['TotalExposure'] = asthma_disease_df[exposure_cols].sum(axis=1)
```
